```python
import pandas as pd
```

```python
df_result = pd.read_csv('socio_results_logit.csv')
df_result
```

```python
columns = ['sex', 'relation', 'is_parent', 'has_pets', 'age', 'mean_tr', 'median_tr', 'tr_per_month']
colums_for_code = ['sex', 'relation', 'is_parent', 'has_pets']
columns_for_disc = ['age', 'mean_tr', 'median_tr', 'tr_per_month']
```

```python
df_result['accuracy_mean'] = df_result[colums_for_code].mean(axis=1)
df_result['accuracy_nrmse'] = df_result[columns_for_disc].mean(axis=1)
```
